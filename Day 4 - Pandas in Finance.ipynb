## Merging Dataframes

```python
import pandas as pd

# IBM
df_ibm = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/IBM?period1=27043200&period2=1605047821&interval=1d&events=history&includeAdjustedClose=true')

# AAPL
df_aapl = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=27043200&period2=1605047821&interval=1d&events=history&includeAdjustedClose=true')

# AMZN
df_amzn = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/AMZN?period1=27043200&period2=1605047821&interval=1d&events=history&includeAdjustedClose=true')
```

```python
# Clean up the aapl dataframe before merging
df2_aapl = df_aapl.loc[:, ['Date', 'Adj Close']]
df2_aapl
```

```python
# Clean up the ibm dataframe before merging
df2_ibm = df_ibm.loc[:, ['Date', 'Adj Close']]
df2_ibm
```

```python
df_all = pd.merge(df2_ibm, df2_aapl, on = 'Date')
df_all
```

```python
df_all_outer = pd.merge(df2_ibm, df2_aapl, on = 'Date', how='outer')
df_all_outer
```

```python
# Add column name suffixes (note the space)
df_all = pd.merge(df2_ibm, df2_aapl, on = 'Date', suffixes=[' IBM', ' AAPL'])
df_all
```

```python
# Outer merge 
df_all = df2_ibm.merge(df2_aapl, on = 'Date', suffixes=[' IBM', ' AAPL'], how='outer')
df_all
```

```python
# Merge in AMZN
df2_amzn = df_amzn.loc[:, ['Date', 'Adj Close']]
df_all2 = pd.merge(df_all, df2_amzn, on = 'Date', how = 'outer')
df_all2
```

```python
df_all2 = df_all2.rename(columns={'Adj Close':'Adj Close AMZN'})
df_all2
```

```python
df_all3 = df_all2.dropna()
df_all3
```

```python
# Faster way of merging several Dataframes
dfs = [
    df2_ibm.set_index('Date').rename(columns={'Adj Close':'IBM'}), 
    df2_aapl.set_index('Date').rename(columns={'Adj Close':'AAPL'}), 
    df2_amzn.set_index('Date').rename(columns={'Adj Close':'AMZN'})
    ]

# Don't forget axis=1
df_final = pd.concat(dfs, join='outer', axis=1)
df_final
```

## Visualization

```python
df_stocks = df_final.dropna()
df_stocks
```

```python
# Normalize the prices
def normalize_price(x):
    return x / x[0]

df_display = df_stocks.apply(normalize_price)
df_display
```

```python
# Size
df_display.plot(kind='line', title='Stock Prices', logy=True, grid=True, figsize=(15, 8))
```

```python
# Get the returns for AAPL
df_aapl_returns = df_stocks.loc[:, ['AAPL']]
df_aapl_returns['Return'] = df_aapl_returns['AAPL'].pct_change()
df_aapl_returns
```

```python
df_aapl_returns.loc[:, ['Return']]
```

```python
df_aapl_returns['Return'].plot(kind='hist', bins=200)
```

```python
# Blank cells can cause issues with rolling()
df_aapl = df_aapl.dropna()
df_aapl
```

```python
df_aapl['5 Day Moving Average'] = df_aapl['Adj Close'].rolling(5).mean()
df_aapl.head(10)
```

```python
df_aapl['200 Day Moving Average'] = df_aapl['Adj Close'].rolling(200).mean()
df_aapl.head(200)
```

```python
df_aapl = df_aapl.set_index('Date')
df_aapl
```

```python
df_aapl.loc[:, ['Adj Close', '5 Day Moving Average', '200 Day Moving Average']].plot(kind='line', figsize=(17, 7), title='Apple (AAPL)')

df_aapl.loc[:, ['Volume']].plot(kind='area', color='red', figsize=(17, 3))
```
